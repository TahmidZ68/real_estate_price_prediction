# #Train test

```python
import numpy as np
def split_train_test(data,test_ratio):
    shuff = np.random.permutation(len(data))
    test_set_size = int(len(data)*test_ratio)
    test_ind = shuff[:test_set_size]
    train_ind =shuff[test_set_size:]
    return data.iloc[train_ind],data.iloc[test_ind]
    
```

```python
train_set, test_set = split_train_test(df, 0.2)
```

```python
print("Rows in train set: ",len(train_set))
```

```python
print("Rows in test set: ",len(test_set))
```

# #Shuffle_Split

```python
 from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)
for train_index, test_index in split.split(df,df['CHAS']):
    strat_train_set = df.loc[train_index]
    strat_test_set = df.loc[test_index]
```
